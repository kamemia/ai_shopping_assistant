# Changing Data Types of Columns

```python
characters_remove = [',','₹','%']
columns = ['discounted_price','actual_price','discount_percentage','rating_count']
for ch in characters_remove:
    for cl in columns:
        df[cl] = df[cl].str.replace(ch,'')

#changetype now
for cl in columns:
    df[cl] = df[cl].astype('float64')
    
```

# Total sales for each category

```python
df['sales'] =  df.discounted_price*df.rating_count
df.sales.head()
```

```python
total_sales  = df.groupby('category')['sales'].sum().sort_values(ascending = False)
```

> **Math Function to convert Numbers into readable formats**

```python
import math

millnames = ['',' Thousand',' Million',' Billion',' Trillion']

def millify(n):
    n = float(n)
    millidx = max(0,min(len(millnames)-1,
                        int(math.floor(0 if n == 0 else math.log10(abs(n))/3))))

    return '{:.0f}{}'.format(n / 10**(3 * millidx), millnames[millidx])
```

```python
values = []
for x in total_sales.values:
    values.append(millify(x))
    print(millify(x))
```

```python
fig, ax = plt.subplots(figsize=(14, 5))

ax.set_title('Total Sales of Categories',fontweight='bold' )
sns.barplot(ax= ax,y =total_sales.index,x =total_sales.values,palette='Blues_r')
ax.set_ylabel(' Categories', fontsize = 20)
ax.bar_label(ax.containers[0], labels=values, padding=3)

```

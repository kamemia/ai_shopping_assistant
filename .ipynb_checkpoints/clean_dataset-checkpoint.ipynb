```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv("data/amazon_products.csv")
df
```

For this task, we will only be needing specific columns

```python
df.shape
```

```python
df.info()
```

```python
df_subset = df[["Product_Name", "Category" ,"Product Url", "Selling Price"]]
```

```python
df_subset
```

```python
print(df_subset.isnull().sum())
```

```python
clean_df = df_subset.dropna()
```

```python
print(clean_df.isnull().sum())
```

```python
clean_df.head(10)
```

```python
# new data frame with split value columns. We use n = 3 to get a maximum of 3+1 columns
new = clean_df["Category"].str.split("|", n = 3, expand = True)
  
# making the first category called Main Category
clean_df["product_category"]= new[0] 
# Dropping old category columns 
clean_df.drop(columns =["Category"], inplace = True)
```

```python
clean_df.head()
```

```python
dataset = clean_df.rename(columns={'Product_Name': 'product_name', 'Product Url': 'product_url', 'Selling Price': 'product_price'})
```

```python
dataset.head(3)
```

```python
dataset.to_csv('data/amazon.csv', index=False)
```
